```python
import random
import json
import numpy as np
```

```python
import nltk
nltk.download("all")
from nltk.stem import WordNetLemmatizer
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
```

Let's Initializing the lemmatizer to convert all words in the sentence to a single understandable
word.

```python
lemmatizer = WordNetLemmatizer()
```

load the data

```python
intents = json.loads(open("LAB01_AI-ETHICS_Intent.json").read())
```

```python
words = []
classes = []
documents = []
ignore_letters = ["?", "!", ".", ","]
```

Part of Feature Extracting:- We Extract features from the data and appending it into different classes corresponding to
training and testing data and while extracting the data tokenizing the sentence.

```python
# for intent in intents["intents"]:
#     for pattern in intent["text"]:
#         word_list = nltk.word_tokenize(pattern)
#         words.extend(word_list)
#         documents.append((word_list, intent["tag"]))
#         if intent["tag"] not in classes:
#             classes.append(intent["tag"])
                      
for intent in intents['intents']:
    for pattern in intent['text']:
        # tokenize each word in the sentence
        word_list = nltk.word_tokenize(pattern)
        # add to our words list
        words.extend(word_list)
        # add to documents in our corpus
        documents.append((word_list, intent["text"]))
        # add to our classes list
        if intent['text'] not in classes:
            classes.append(intent['text'])
```

Lemmatize each word in the words list

```python
words = [lemmatizer.lemmatize(word) for word in words if word not in ignore_letters]
words = sorted(set(words))
```

```python
print(classes)
```

```python
classes = sorted(set(classes))
```

```python
training = []
output_empty = [0] * len(classes)
```

```python
for document in documents:
    bag = []
    word_patterns = document[0]
    word_patterns = [lemmatizer.lemmatize(word.lower()) for word in word_patterns]
    for word in words:
        if word in word_patterns:
            bag.append(1)
        else:
            bag.append(0)
    output_row = list(output_empty)
    output_row[classes.index(document[1])] = 1
    training.append([bag, output_row])
```

```python
random.shuffle(training)
```

```python
training = np.array(training)
```

```python
train_x = list(training[:, 0])
train_y = list(training[:, 1])
```

Initiliazing the Sequential Model for training

```python
model = Sequential()
```

Defining the Neural Network for training

```python
model.add(Dense(128, input_shape=(len(train_x[0]),), activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation="softmax"))
```

```python
sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=["accuracy"])
```

Training the model using the data that we clustered

```python
model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)
```

Allowing the same process when user inputs a data, same data preprocessing as we
required for training the model


```python
def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [lemmatizer.lemmatize(word) for word in sentence_words]
    return sentence_words
```

```python
def bag_of_words(sentence):
    sentence_words = clean_up_sentence(sentence)
    bag = [0] * len(words)
    for w in sentence_words:
        for i, word in enumerate(words):
            if word == w:
                bag[i] = 1

    return np.array(bag)
```

```python
def predict_class(sentence):
    bow = bag_of_words(sentence)
    res = model.predict(np.array([bow]))[0]
    ERROR_THRESHOLD = 0.25
    results = [[i, r] for i, r in enumerate(res) if r > ERROR_THRESHOLD]

    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": classes[r[0]], "probability": str(r[1])})

    return return_list
```

```python
def get_response(intents_list, intents_json):
    tag = intents_list[0]["intent"]
    list_of_intents = intents_json["intents"]
    for i in list_of_intents:
        if i["text"] == tag:
            result = random.choice(i["responses"])
            break
    return result
```

Running the bot

```python
print("Bot is running!")
while True:
    message = input()
    ints = predict_class(message)
    res = get_response(ints, intents)
    print("Bot Response: ", res)
```
